```python
# Importing libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

#### 1. Creating Data Frame

```python
# Creating Data Frame

path = r"C:\Users\David\Desktop\CareerFoundry\Achievement 4\Instacart Basket Analysis"

df = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'instacart_data.pkl'))
```

#### 2. Bar Plot - Order Day of Week

```python
# Creating bar plot of 'order_dow' column

df.order_dow.value_counts().sort_index().plot.bar()

# Saving bar plot to .png file
# plt.savefig(os.path.join(path, '04 Analysis', 'Visualizations','order_dow_freq.png'))
```

#### 3. Histogram - Prices

```python
# Creating histogram plot of 'prices' column

df.prices.plot.hist(bins=25)


# Saving histogram to .png file
# plt.savefig(os.path.join(path, '04 Analysis', 'Visualizations','prices_freq.png'))
```

#### 4. Scatterplot - Prices

```python
# Creating scatterplot of 'prices' column

sns.scatterplot(data = df, x='prices', y='prices')

# Saving scatterplot to .png file
# plt.savefig(os.path.join(path, '04 Analysis', 'Visualizations','prices_scatter.png'))
```

#### 5. Creating subsets

```python
# Creating a random sample

np.random.seed(7)
dev = np.random.rand(len(df)) <= 0.7
```

```python
# Splitting data frame into subsets

big = df[dev]
small = df[~dev]
```

#### 6. Lineplot - Order Day of Week & Prices

```python
# Creating df_2 based on small small subset

df_2 = small[['order_dow', 'prices']]
```

```python
# Creating lineplot using df_2 to compare the days of the week (order_dow column) to the 'prices' column

sns.lineplot(data = df_2, x = 'order_dow', y = 'prices')

# Saving line plot to .png file
# plt.savefig(os.path.join(path, '04 Analysis', 'Visualizations','order_dow_line.png'))
plt.show()
```

#### 7. Histogram - Order Hour of Day

```python
# Creating histogram of order_hour_of_day column

df.order_hour_of_day.plot.hist(bins=24, figsize=(12,8), xticks=([num for num in range(0,24)]))
plt.ticks=[num for num in range(0,24)]
```

```python
# Setting customization parameters for figure colors
sns.set(rc={'axes.facecolor': '#fcfdfc', 'figure.facecolor':'#fcfdfc'})

# Creating histogram of order_hour_of_day column
plt.figure(figsize=(12,8))
h = sns.histplot(df['order_hour_of_day'], bins=24, color='#60ab59')
plt.xlabel('Hour of Day')
plt.ylabel('Frequency')
plt.title('Frequncy of Orders by Hour of Day')

h.set_xticks([num for num in range(0,24)])
h.set_xticklabels([num for num in range(0,24)])


# Saving histogram to .png file
#plt.savefig(os.path.join(path, '04 Analysis', 'Visualizations','order_hour_of_day_hist.png'))
plt.show()
```

The histogram plot above shows that the distribution of orders is fairly symmetrical overall, with a smaller tail on the left of the distribution. This tells us that most orders are placed in the middle of the day between 9am and 4pm, with a peak number of orders occuring between 9am and 10am.

#### 8. Bar Plot - Loyalty Flag

```python
# Creating bar plot from "loyalty_flag" column

plt.figure(figsize = (12,8))
df.loyalty_flag.value_counts().sort_values(ascending=False).plot.bar(color='#60ab59', rot=0)
plt.xlabel('Customer Type')
plt.ylabel('Frequency')
plt.title('Frequency of Orders by Loyalty Type')


# Saving barplot to .png file
# plt.savefig(os.path.join(path, '04 Analysis', 'Visualizations','loyalty_orders.png'))
plt.show()

```

#### 9. Line Plot - Order Hour of Day & Prices

```python
# Creating df_3 from small subset to compare 'order_hour_of_day', 'prices', and 'gender' columns

df_3 = small[['order_hour_of_day', 'prices', 'gender']]
```

```python
# Creating a lineplot based on df_3 subset to compare 'order_hour_of_day' and 'prices' column

plt.figure(figsize = (12,8))
sns.lineplot(data = df_3, x = 'order_hour_of_day', y = 'prices', color = '#60ab59')
plt.xticks([num for num in range(0,24)])
plt.xlabel('Hour of Day')
plt.ylabel('Prices')
plt.title('Total Spend by Hour of Day')

# Saving lineplot to .png file
# plt.savefig(os.path.join(path, '04 Analysis', 'Visualizations','hour_prices_line.png'))
plt.show()
```

Orders with a higher overall price are placed earlier in the day, specifically between 4am and 5am. This indicates that while most orders are placed in the afternoon, high value orders (possibly for business' or restaurants for example) are placed much earlier in the day.

```python
# Adding a hue to the above lineplot to add 'gender' as a paramater

plt.figure(figsize = (12,8))
sns.lineplot(data = df_3, x = 'order_hour_of_day', y = 'prices', hue='gender')
plt.xticks([num for num in range(0,24)])
plt.xlabel('Hour of Day')
plt.ylabel('Prices')
plt.title('Total Spend by Hour of Day & Gender')

# Saving lineplot to .png file
# plt.savefig(os.path.join(path, '04 Analysis', 'Visualizations','hour_prices_line_gender.png'))
plt.show()
```

Using gender to distinguish between orders and the hour of the day we can see that men are generally placing these higher value orders earlier in the day, specifically at 1am and 4am.

#### 10. Line Plot - Age & Dependants

```python
# Creating df_4 from small subset to compare 'age' and 'dependants' columns

df_4 = small[['age', 'dependants']]
```

```python
# Creating a line plot to compare relationship between age and number of dependants

plt.figure(figsize = (12,8))
sns.lineplot(data = df_4, x = 'age', y = 'dependants', color='#60ab59')
plt.xlabel('Age')
plt.ylabel('Dependants')
plt.title('Relationship Between Age and Dependants')

# Saving line plot to .png
# plt.savefig(os.path.join(path, '04 Analysis', 'Visualizations','age_dependants_line.png'))
plt.show()
```

There is no clear correlation between the age of a customer and the number of dependants.

#### 11. Scatterplot - Age & Income

```python
# Creating scatterplot to visualize relationship between age and income

plt.figure(figsize = (12,8))
sns.scatterplot(data=df, x='age', y='income', color='#60ab59')
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Relationship Between Age and Spending Power')
# plt.savefig(os.path.join(path, '04 Analysis', 'Visualizations','age_income_scatter.png'))
plt.show()
```

Most data points fall below the 200,000 range. All customers with over $400,000 in spending power are over the age of 40. When determining spending power on a more granular level it may be helpful to to create age categories (under/over 40 for example) to lower the number of data points and look for any further correlation.
