# Combining & Exporting Data

## Steps in Script:
#### 1. Creating Data Frames
#### 2. Cleaning columns in imported Data Frames
#### 3. Merging Data Frames
#### 4. Exporting to file

```python
# Importing libraries

import pandas as pd
import numpy as np
import os

path = r"C:\Users\David\Desktop\CareerFoundry\Achievement 4\Instacart Basket Analysis"
```

#### 1. Creating Data Frames

```python
df_op_combined = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_combined.pkl'))
df_prods = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'products_clean.csv'))
```

```python
df_op_combined.head()
```

#### 2. Cleaning Columns in Imported Data Frames

```python
df_op_combined = df_op_combined[df_op_combined.columns.to_list()[:-1]]
df_op_combined.shape
```

```python
df_prods = df_prods[df_prods.columns.to_list()[1:]]
```

```python
df_prods.head()
```

#### 3. Merging Data Frames

```python
orders_products_merged = df_op_combined.merge(df_prods, on='product_id', indicator=True)
```

```python
orders_products_merged.head()
```

#### 4. Exporting to file

```python
orders_products_merged.to_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'order_products_merged.pkl'))
```
