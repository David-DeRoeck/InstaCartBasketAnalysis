# Creating a Department ID search for Subsetting Data Frames

```python
def dept_search():
    search = False
    for item in new_data_dict.values():
        print(f'--{item.title()}--')
    while search == False:
        items = input('Which aisle? ').lower()
        if items in new_data_dict.values():
            for value, item in new_data_dict.items():
                if items == item:
                    aisle = int(value)
                    search = True
        else:
            print("Sorry, that aisle is not available to search.")
    return df_prods[df_prods.department_id == aisle]
```
